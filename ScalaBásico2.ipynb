### Documentación de Archivos de Datos de IMDb

Como datasets utilizaremos una version reducida de [IMDb Non-Commercial Datasets](https://developer.imdb.com/non-commercial-datasets/)


__title.akas.tsv.gz__

- **titleId (string)**: un `tconst`, un identificador único alfanumérico del título.
- **ordering (integer)**: un número para identificar de manera única las filas para un `titleId` dado.
- **title (string)**: el título localizado.
- **region (string)**: la región para esta versión del título.
- **language (string)**: el idioma del título.
- **types (array)**: Conjunto enumerado de atributos para este título alternativo. Uno o más de los siguientes: `"alternative"`, `"dvd"`, `"festival"`, `"tv"`, `"video"`, `"working"`, `"original"`, `"imdbDisplay"`. Pueden agregarse nuevos valores en el futuro sin previo aviso.
- **attributes (array)**: Términos adicionales para describir este título alternativo, no enumerados.
- **isOriginalTitle (boolean)**: `0`: no es el título original; `1`: es el título original.

__title.basics.tsv.gz__

- **tconst (string)**: identificador único alfanumérico del título.
- **titleType (string)**: el tipo/formato del título (por ejemplo, `movie`, `short`, `tvseries`, `tvepisode`, `video`, etc.).
- **primaryTitle (string)**: el título más popular / el título utilizado por los cineastas en materiales promocionales en el momento del lanzamiento.
- **originalTitle (string)**: título original, en el idioma original.
- **isAdult (boolean)**: `0`: título no para adultos; `1`: título para adultos.
- **startYear (YYYY)**: representa el año de lanzamiento de un título. En el caso de series de TV, es el año de inicio de la serie.
- **endYear (YYYY)**: año de finalización de la serie de TV. `'\N'` para todos los demás tipos de títulos.
- **runtimeMinutes**: duración principal del título, en minutos.
- **genres (string array)**: incluye hasta tres géneros asociados con el título.

__title.crew.tsv.gz__

- **tconst (string)**: identificador único alfanumérico del título.
- **directors (array of nconsts)**: director(es) del título dado.
- **writers (array of nconsts)**: escritor(es) del título dado.

__title.episode.tsv.gz__

- **tconst (string)**: identificador alfanumérico del episodio.
- **parentTconst (string)**: identificador alfanumérico de la serie de TV principal.
- **seasonNumber (integer)**: número de temporada a la que pertenece el episodio.
- **episodeNumber (integer)**: número de episodio del `tconst` en la serie de TV.

__title.principals.tsv.gz__

- **tconst (string)**: identificador único alfanumérico del título.
- **ordering (integer)**: un número para identificar de manera única las filas para un `titleId` dado.
- **nconst (string)**: identificador único alfanumérico del nombre/persona.
- **category (string)**: la categoría del trabajo que desempeñó esa persona.
- **job (string)**: el título específico del trabajo si es aplicable, de lo contrario `'\N'`.
- **characters (string)**: el nombre del personaje interpretado si es aplicable, de lo contrario `'\N'`.

__title.ratings.tsv.gz__

- **tconst (string)**: identificador único alfanumérico del título.
- **averageRating**: promedio ponderado de todas las calificaciones individuales de los usuarios.
- **numVotes**: número de votos que ha recibido el título.

__name.basics.tsv.gz__

- **nconst (string)**: identificador único alfanumérico del nombre/persona.
- **primaryName (string)**: nombre con el que la persona es más conocida.
- **birthYear**: en formato `YYYY`.
- **deathYear**: en formato `YYYY` si aplica, de lo contrario `'\N'`.
- **primaryProfession (array of strings)**: las tres principales profesiones de la persona.
- **knownForTitles (array of tconsts)**: títulos por los que la persona es conocida.

### Ejercicio 1.1: Creación de un Objeto
Crea un objeto llamado IMDBStatsObject que debe tener los siguientes métodos:

-anioLanzamientoPelicula: Devuelve el año de lanzamiento de una película dado un título y un DataFrame con el esquema title.basic.

-promedioRatingEntreNumVotos: Dado un DataFrame con un esquema title.ratings, devuelve uno nuevo con una columna llamada "nuevaColumna" en la que se indica el promedio del rating dividido entre el número de votos.


```scala
// Creo esta celda para leer todos los df

val title_basics = spark.read.option("header", "true")
                    .option("sep", "\t")
                    .option("inferSchema", "true")
                    .csv("Files/title_basics.csv")

val title_ratings = spark.read.option("header", "true")
                    .option("sep", "\t")
                    .option("inferSchema", "true")
                    .csv("Files/title_ratings.csv")
```

```scala
import org.apache.spark.sql.DataFrame
import org.apache.spark.sql.functions.col // Importamos lo necesario para el código


object IMBDStatsObject {
    // Creamos el método para devolver el año de lanzamiento de la película
    def anioLanzamientoPelicula(titulo: String, df: DataFrame): Unit = {

        // Filtramos para seleccionar el año de lanzamiento de la película que se introduzca
        val anio = df.filter(col("primaryTitle") === titulo).select("startYear").first().get(0)

        // Mostramos el año por pantalla
        println(s"El año de lanzamiento de $titulo es $anio")
    
    }

    // Creamos el método para devolver el dataframe con la nueva columna
    def promedioRatingEntreNumVotos(df: DataFrame): DataFrame = {
        // Añadimos la nueva columna que divide el promedio del rating entre el número total de votos
        df.withColumn("nuevaColumna", col("averageRating")/col("numVotes"))

    }

}


// Ejemplo método 1: probamos con la película "Baby's Meal"
IMBDStatsObject.anioLanzamientoPelicula("Baby's Meal", title_basics)


// Ejemplo método 2: ejecutamos con el dataframe title_ratings
val promedioRating = IMBDStatsObject.promedioRatingEntreNumVotos(title_ratings)

promedioRating.show(10)
```

### Ejercicio 1.2: Creación de una Clase
Crea un objeto llamado MovieSorter para analizar los ratings de las películas. Los métodos deben incluir:

- getIdMejorValorada: Devuelve el identificador de la película con la calificación más alta.
- getIdPeorValorada: Devuelve el identificador de la película con la calificación más baja.
- getEnRangoValor: Devuelve las filas que tienen el rating dentro de un rango dado. Probaremos con un rango entre 8 y 10.


```scala

// Pone que cree un objeto, pero deduzco inteligentísimamente que se refiere a que cree una clase

class MovieSorter {
    // Creamos el primer método para devolver el ID de la película mejor valorada
    def getIdMejorValorada(df: DataFrame): Unit = {
        // Objetemos el ID de la película mejor valorada
        val max = df.orderBy(col("averageRating").desc).select("tconst").first.get(0)
        // Mostramos el ID de dicha película
        println(s"El ID de la película mejor valorada es: $max")
    }

    // Creamos el segundo método para devolver el ID la película peor valorada
    def getIdPeorValorada(df: DataFrame): Unit = {
        // Objetemos el ID de la película peor valorada
        val min = df.orderBy(col("averageRating").asc).select("tconst").first.get(0)
        // Mostramos el ID de dicha película
        println(s"El ID de la película mejor valorada es: $min")
    }

    // Creamos el tercer método, para devolver el DataFrame con las filas cuyo rating está entre 8 y 10
    def getEnRangoValor(df: DataFrame): DataFrame = {
        // Filtramos las películas
        df.filter(col("averageRating").between(8,10))
    }
}

// Ejemplo: Instanciamos la clase y llamamos a cada uno de los tres métodos
val ratings = new MovieSorter()
ratings.getIdMejorValorada(title_ratings)
ratings.getIdPeorValorada(title_ratings)
ratings.getEnRangoValor(title_ratings).show()
```
